```python
interests = [
    (0, "Hadoop"), (0, "Big Data"), (0, "HBase"), (0, "Java"),
    (0, "Spark"), (0, "Storm"), (0, "Cassandra"),
    (1, "NoSQL"), (1, "MongoDB"), (1, "Cassandra"), (1, "HBase"),
    (1, "Postgres"), (2, "Python"), (2, "scikit-learn"), (2, "scipy"),
    (2, "numpy"), (2, "statsmodels"), (2, "pandas"), (3, "R"), (3, "Python"),
    (3, "statistics"), (3, "regression"), (3, "probability"),
    (4, "machine learning"), (4, "regression"), (4, "decision trees"),
    (4, "libsvm"), (5, "Python"), (5, "R"), (5, "Java"), (5, "C++"),
    (5, "Haskell"), (5, "programming languages"), (6, "statistics"),
    (6, "probability"), (6, "mathematics"), (6, "theory"),
    (7, "machine learning"), (7, "scikit-learn"), (7, "Mahout"),
    (7, "neural networks"), (8, "neural networks"), (8, "deep learning"),
    (8, "Big Data"), (8, "artificial intelligence"), (9, "Hadoop"),
    (9, "Java"), (9, "MapReduce"), (9, "Big Data")
]

def data_scientists_who_like(target_interest):
    """Find the ids of all users who like the target interest."""
    return [user_id
            for user_id, user_interest in interests
            if user_interest == target_interest]

from collections import defaultdict

# Keys are interests, values are lists of user_ids with that interest
user_ids_by_interest = defaultdict(list)

global_user_ids_by_interest = {}

for user_id, interest in interests:
    """
    interests = [
    (0, "Hadoop"), (0, "Big Data")...(9, "MapReduce"), (9, "Big Data")
    ]
    for user_id [the first value in tuple], interest [the second value is the interest] in the interests list of tuples
    """
    user_ids_by_interest[interest].append(user_id)
    """
    user_ids_by_interest[interest] is the defaultdict(list) for each interest and we append the user_id to each interest
    in the format of: {'Hadoop': [0, 9], 'Big Data': [0, 8, 9]}
    so, the defaultdict(list) must contain all interests, then we are appending the user_id's as they are associated with
    that interest
    """
    global_user_ids_by_interest.update(user_ids_by_interest)

print(global_user_ids_by_interest)
```
